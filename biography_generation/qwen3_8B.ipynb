```python
# mlx imports  # Use to work with MLX, using quantized (4bit) models
import mlx.core as mx
from utils.mlx_utils import load_model, generate_biography

# transformers imports  # Use to work with torch, using non-quantized models
#import torch
#from huggingface_hub import login
#from utils.torch_utils import load_model, generate_biography

# common imports
from config import QWEN3, QWEN3_PARAMS
from utils.common import save_response
from prompts import PROMPTS
```

```python
# TO RUN ONLY FOR TRANSFORMERS
#TOKEN = ""
#login(TOKEN)
```

```python
# Selection of prompt used
PROMPT_VERSION = "v3"
prompt_set = PROMPTS[PROMPT_VERSION]
system_prompt = prompt_set["system"]
user_prompt_nodis = prompt_set["no_dis"]
user_prompt_withdis = prompt_set["with_dis"]

# Seed for reproducibility
SEED = 42

# mxl:
mx.random.seed(SEED)

# transformers:
#torch.manual_seed = SEED

# Load model
model, tokenizer = load_model(QWEN3)
```

```python
# Generate no-disability prompt biography
biography_no_dis = generate_biography(
    model = model,
    tokenizer = tokenizer,
    user_prompt = user_prompt_nodis,
    system_prompt = system_prompt,
    params = QWEN3_PARAMS,
    enable_thinking = False
)
```

```python
# Generate with-disability prompt biography
biography_with_dis = generate_biography(
    model = model,
    tokenizer = tokenizer,
    user_prompt = user_prompt_withdis,
    system_prompt = system_prompt,
    params = QWEN3_PARAMS,
    enable_thinking = False
)
```

```python
# Save results
save_response(data=biography_no_dis,
              filepath="outputs/results_qwen.json",
              prompt_set=PROMPT_VERSION,
              variant="nodis",
              system_prompt=system_prompt,
              user_prompt=user_prompt_nodis)

save_response(data=biography_with_dis,
              filepath="outputs/results_qwen.json",
              prompt_set=PROMPT_VERSION,
              variant="withdis",
              system_prompt=system_prompt,
              user_prompt=user_prompt_withdis)
```
