```python
import mlx.core as mx

from config import MISTRAL, MISTRAL_PARAMS
from utils.mlx_utils import load_model, generate_biography # Use to work with MLX, using quantized (4bit) models
# from utils.torch_utils import load_model, generate_biography # Use to work with torch, using non-quantized models
from utils.common import save_response
from prompts import SYSTEM_PROMPT, PROMPT_NO_DIS, PROMPT_WITH_DIS, build_mistral_prompt
```

```python
# Seed for reproducibility
mx.random.seed(42)

# Load model
model, tokenizer = load_model(MISTRAL)
```

```python
# Generate no-disability prompt biography
biography_no_dis = generate_biography(
    model = model,
    tokenizer = tokenizer,
    user_prompt = build_mistral_prompt(PROMPT_NO_DIS),
    params=MISTRAL_PARAMS
)
```

```python
# Generate with-disability prompt biography
biography_with_dis = generate_biography(
    model = model,
    tokenizer = tokenizer,
    user_prompt = build_mistral_prompt(PROMPT_WITH_DIS),
    params=MISTRAL_PARAMS
)
```

```python
# Save results
save_response(biography_no_dis, "outputs/response_no_dis_mistral_1.json")
save_response(biography_with_dis, "outputs/response_with_dis_mistral_1.json")
```
